# General information

## Course summary

Knowledge is encoded in text and the web is a valuable source of knowledge. This course aims to extract and discover knowledge from online sources of information such us text corpora (e. g., Project Gutenberg, 46,000+ free ebooks) or the social web (e. g., Twitter, 500+ million registered users) by web mining and text mining in Python. The goals of this course are to understand human knowledge as a product of social interactions, to be aware of the value of the web as a source of information, and to be able to extract valuable knowledge from text corpora and the social web.

## Justification

Humanists seek new knowledge to answer the big questions that concern the humanity: *"What is the meaning of life?"*, *"What do I know?"*, *"What is true?"*. They aim to expand the knowledge of the world by expanding their own first. In the third Humanist Manifesto, *Humanism and Its Aspirations - Humanist Manifesto III, a successor to the Humanist Manifesto of 1933*, humanists assert that *"knowledge of the world is derived by observation, experimentation, and rational analysis"*. This course is not intended to answer such major questions but to provide some tools to obtain reliable knowledge that help us to understand our world.

This course assumes the theses of social epistemologists, who argue against the idea that knowledge is produced by individuals. They claim that most of our knowledge is created and transmitted through social processes, and the importance of paying attention to these processes work in order to understand the nature of knowledge. These ideas are consistent with Ikujiro Nonaka and Hirotaka Takeuchi, two of the most experienced authors in knowledge management. Nonaka and Takeuchi propose in their work titled *The knowledge-creating company: how Japanese companies create the dynamics of innovation* a model whereby the creation of knowledge is originated by the individual and, again through social processes, finalized by the collective.

These social interactions are coming to be evident in the academia and the research community, where interdisciplinary collaboration is becoming increasingly important to face big challenges. But besides understanding the social nature of knowledge itself, it is also essential to do it in a context in which technology and new forms of media, new distribution channels and new delivery systems are affecting traditional means of creating and disseminating knowledge. In particular, social networks and knowledge communities are specially reshaping the way knowledge is constructed by taking advantage of crowdsourcing. Hence, the importance of understanding how networks work and how knowledge emerges from their members' behaviour.

As information contained on social networks and online communities are generated by human beings, this is especially valuable for discovering knowledge about people's opinions and preferences, in addition to many other kinds of knowledge that we encode in text. Therefore, providing a strong background in text processing is fundamental.

Social networks involve disciplines like sociology, mathematics, and computer science, whereas text processing involves linguistics, statistics, and artificial intelligence. They are two good examples of interdisciplinary fields. For this reason, students from all backgrounds, but especially from the humanities, social science, computer science and engineering, are strongly encouraged to take this course.

## Methodology

This course is formed by 26 lectures over 13 weeks (two sessions per week). Each session is 1-hour long. For the first 20 minutes of each class, I will explain theoretical concepts and demonstrate the use of particular programming constructs using notebooks that I supply. The next 40 minutes will be used for student hands-on exercises, where they will practice programming in class while I answer their questions and provide assistance. Students will also share and discuss their own solutions in terms of clarity, efficiency, and other criteria.

The course aims to discover objective knowledge from online resources that supports future decisions by means of programming exercises focused on social media mining and natural language processing. To achieve these goals, the contents are divided into 5 modules:

1. Knowledge. This is an introduction to the course. It introduces some concepts on knowledge, emphasizing on its social nature.

2. Python. This module is a programming guide with Python that develops all the programming skills needed to face the rest of the course.

3. Social web. This part of the course provides useful techniques to mine social media. It shows, among others, how to get a user's friend network, the most common words used by a community or search for trending topics by specific locations.

4. Text mining. This module covers subjects related to text manipulation, resource accessing and applied natural language processing like word tagging and categorization, or document classification.

5. Projects. Time for the students to work on their assignments, ask questions, and present their final projects.

All lectures are delivered in the form of IPython Notebook and available through IPython Notebook viewer. They combine brief description of the theory with practical exercise to apply the concepts explained. The contents are taught in a broad and general way, but the examples and the assignments have been carefully designed to work with texts and contents on social media, as these are the main subjects of study in this course. Students will work on their computers.

## Teaching philosophy

I have been a Computer Science lecturer and a Spanish instructor. From these very different experiences, I have drawn two conclusions:

1. Students are more productive when they enjoy their lectures.

2. Each student has his/her own process to acquire knowledge and develop his/her own inner skills.

To make lectures enjoyable, in-class learning must be a gratifying experience. I pride myself to create a safe learning environment. To achieve this, I promote freedom for my students to openly voice their questions or comments in the classroom. I treat all their ideas as opportunities for everyone to learn, I give them appropriate and respectful responses, and I celebrate good contributions. At the same time, I use the classroom space to move around and interact with my students. I do my best to be perceived as an approachable person and model kindness every chance I get. I believe that my role in class must be to help them rather than to evaluate them. My formal evaluations as an instructor suggest that they perceive me as a friendly teacher, who creates a fun, comfortable, and stimulating learning atmosphere, and who portrays a committed attitude.

Based on the fact that each student has their own necessities and personal background and skills, my role as their instructor involves listening to them, understanding their needs, and working to meet these needs for each individual.

I do not think that classic memorization is an effective way to learn. Instead, I prefer that my students learn by doing. For this reason, my lectures are focused, as much as possible, on the practical aspects of the topics taught. Inevitably, there always are lectures with a stronger theoretical viewpoint. But in general, I briefly describe the theory and then we see real applications of the concepts learned, giving a sense of utility to my lectures.

For the same reason, I do not evaluate my courses via exams, but rather by working on practical assignments. Rather, I evaluate their aptitude to solve assignments by allowing them to utilize all resources at hand. These resources vary from their own intellectual capabilities to resources in libraries or online instruments and their grade of involvement with the activities of course.


# Syllabus

## Course description

This course aims to teach how to obtain valuable knowledge from a variety of online sources of information. It starts with an introduction to knowledge and ways of creating it, including how technology and new media are affecting traditional means of its production and dissemination. In particular, the course examines the social nature of knowledge, that claims that human knowledge is created and transmitted through social processes, against more traditional theories of knowledge that argue that it is produced by individuals.

In this context of collectivity, the course provides an overview of crowdsourcing (that is, the act of outsourcing a task to the crowd), collaboration and interdisciplinarity. It also outlines some of the existing mechanisms to encourage the community to participate, as well as the role of reputation and trust in the process of collaborative creation and sharing of knowledge. The exploration of certain online communities formed on Facebook, Twitter or Stack Overflow around specific topics exemplifies these concepts.

The course shows how to extract knowledge from such communities by means of two combined approaches: social media mining and text processing. In order to address these topics, it begins with a general programming guide in Python. Then it covers some useful methods to mine social media like Twitter or Facebook, and websites, and also explores powerful Python libraries to visualize the data extracted like Matplotlib or Pandas. Finally, it showcases major techniques for text processing, namely, information retrieval, text classification or sentiment analysis, which demonstrate it is possible to discover interesting patterns and support decision making from knowledge previously extracted, with an emphasis on practical applications.

## Objectives

The main goals of this course are:

- Understand human knowledge as a social construct, and the specific role of crowdsourcing, interdisciplinary collaboration, and new technologies.

- Be aware of the valuable knowledge contained in new media such as social networks and online knowledge communities.

- Highlight existing mechanisms to identify reliable knowledge, in particular, reputation and trust.

- Create the next generation of skilled researchers by means of programming applied to text processing.

## Learning outcomes

Upon completing this course, students will be able to effectively:

- Write programs in Python in order to algorithmically perform tasks, such as calculating frequency distributions or identifying trending topics on social media.

- Critically judge the validity of online sources of knowledge.

- Access a varied range of online resources to extract information.

- Be able to discover knowledge from online knowledge communities.

## Plagiarism

Plagiarism is a major academic offense. It is the inclusion of someone else's verbatim or paraphrased text in one's own written work without immediate reference. Verbatim text must be surrounded by quotation marks or indented if it is longer than four lines. A reference must follow right after borrowed material (usually the author's name and page number). Without immediate reference to borrowed material, a list of sources at the end of a written assignment does not protect a writer against a possible charge of plagiarism. This also applies to work facilitated or written for students by third parties. Submit all essays to Turnitin. ## Assignments and final project

_The assignments and final project will be submitted via OWL._

**Assignment 1: _Reputation and trust in the social web_**

Individual assignment.

How are reputation and trust affecting the way knowledge is constructed on sites like [StackOverflow](https://stackoverflow.com/), in opposition to sites with no reputation system such as [Wikipedia](https://en.wikipedia.org/)? Why do you think Wikipedia has not implemented a reputation system? How would you implement a reputation system for [Wikipedia](https://en.wikipedia.org/)? How would this system change the way [Wikipedia](https://en.wikipedia.org/) works? Do you think its performance would get better or worse? Why? How would this system deal with trolls?

Write an essay (1,500-3,000 words / 3-6 single-spaced pages) about reputation and trust on [StackOverflow](http://stackoverflow.com/) and the way they affect the community in creating knowledge. Include answers for the previous questions. The article *[Welcome to the new reputation economy](http://www.wired.co.uk/magazine/archive/2012/09/features/welcome-to-the-new-reputation-economy)* ([Wired UK](http://www.wired.co.uk/), 2012) is a highly recommended start point.

Due date: Two weeks after the end of Module I at 11:59pm.

**Assignment 2: _Comparing vocabularies on [Facebook](https://www.facebook.com/) communities_**

Assignment in groups.

Individually, each member of the team:

- Choose your 10 closest friends (or those 10 you most interact with) on [Facebook](https://www.facebook.com/). Your 10 friends and you form a *community*.

- Collect your community's most recent posts.

- Extract the list of all words. Keep duplicate words in the list (in order to count them), normalize capitalization (all words in lowercase), remove punctuation marks, numbers and stop words. Stop words are too frequent words. A list of stop words can be found here:

```“i”, “me”, “my”, “myself”, “we”, “our”, “ours”, “ourselves”, “you”, “your”, “yours”, “yourself”, “yourselves”, “he”, “him”, “his”, “himself”, “she”, “her”, “hers”, “herself”, “it”, “its”, “itself”, “they”, “them”, “their”, “theirs”, “themselves”, “what”, “which”, “who”, “whom”, “this”, “that”, “these”, “those”, “am”, “is”, “are”, “was”, “were”, “be”, “been”, “being”, “have”, “has”, “had”, “having”, “do”, “does”, “did”, “doing”, “a”, “an”, “the”, “and”, “but”, “if”, “or”, “because”, “as”, “until”, “while”, “of”, “at”, “by”, “for”, “with”, “about”, “against”, “between”, “into”, “through”, “during”, “before”, “after”, “above”, “below”, “to”, “from”, “up”, “down”, “in”, “out”, “on”, “off”, “over”, “under”, “again”, “further”, “then”, “once”, “here”, “there”, “when”, “where”, “why”, “how”, “all”, “any”, “both”, “each”, “few”, “more”, “most”, “other”, “some”, “such”, “no”, “nor”, “not”, “only”, “own”, “same”, “so”, “than”, “too”, “very”, “s”, “t”, “can”, “will”, “just”, “don”, “should”, “now”```

- Create a frequency distribution of your words.

- Represent the distribution in a histogram.

With your team:

- Put all the distributions together.

- Contrast them in a histogram.

- Analyze the results: What are the most common words for all the communities? Do your communities have many words in common? Is there any word very frequent in one community but rare in the others? Give reasons to justify the similarities or differences in the vocabulary used by each community.

Due date: Two weeks after the end of Module III at 11:59pm.

**Final project**

Project in groups.

The goal of the final project is to extract knowledge from any of the sources of information seen in the course. The students are free to choose a problem of their own interest, including an appropriate dataset. For the successful completion of this project, they have to:

1. Program the solution of their problem in IPython Notebook (format: file with .ipynb extension and no execution errors).

2. Write a memory research where the students must:

    - Justify the interest and viability of the proposal.
    
    - Explain in what discipline this problem is relevant and provide an example of real application of the solution.
    
    - Explain the kind of disciplinary collaboration by the interaction of the team members' own disciplines.
    
    - Explain the evolution from data to knowledge by means of text mining.
    
    - Describe the solution.
    
    - Explain how the code in Python implements the solution.

3. Present their project in class.

Submission length:

1. Python code: as many code lines as needed.

2. Memory research: 4,000-5,000 words / 8-10 single-spaced pages.

3. Oral presentation: 10-15 minutes (depending on the number of groups).

Due date: Two weeks after the end of Module IV at 11:59pm.
