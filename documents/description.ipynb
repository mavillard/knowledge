###Course description

This course aims to teach how to obtain valuable knowledge from a variety of online sources of information. It starts with an introduction to knowledge and ways of creating it, including how technology and new media are affecting traditional means of its production and dissemination. In particular, the course examines the social nature of knowledge, that claims that human knowledge is created and transmitted through social processes, against more traditional theories of knowledge that argue that it is produced by individuals.

In this context of collectivity, the course provides an overview of crowdsourcing (that is, the act of outsourcing a task to the crowd), collaboration and interdisciplinarity. It also outlines some of the existing mechanisms to incentive the community to participate, as well as the rol of reputation and trust in the process of collaborative creation and sharing of knowledge. The exploration of certain online communities formed on Facebook, Twitter or Stack Overflow around specific topics exemplifies these concepts.

The course shows how to extract knowledge from such communities by means of two combined approaches: social media mining and text processing. In order to address these topics, it begins with a general programming guide in Python. Then it covers some useful methods to mine social media like Twitter or Facebook, and websites, and also explores powerful Python libraries to visualize the data extracted like Matplotlib or Pandas. Finally, it showcases major techniques for text processing, namely, information retrieval, text classification or sentiment analysis, which demonstrate it is possible to discover interesting patterns and support decision making from knowledge previously extracted, with an emphasis on practical applications.

###Justification

Humanists seek new knowledge to answer the big questions that concern the humanity: *"What is the meaning of life?"*, *"What do I know?"*, *"What is true?"*. They aim to expand the knowledge of the world by expanding their own first. In the third Humanist Manifesto, *Humanism and Its Aspirations - Humanist Manifesto III, a successor to the Humanist Manifesto of 1933*, humanists assert that *"knowledge of the world is derived by observation, experimentation, and rational analysis"*. This course is not intended to answer such major questions but to provide some tools to obtain reliable knowledge that help us to understand our world.

This course assumes the theses of social epistemologists, who argue against the idea that knowledge is produced by individuals. They claim that most of our knowledge is created and transmitted through social processes, and the importance of paying attention to these processes work in order to understand the nature of knowledge. These ideas are consistent with Ikujiro Nonaka and Hirotaka Takeuchi's, two of the most experienced authors in knowledge management. Nonaka and Takeuchi propose in their work titled *The knowledge-creating company: how Japanese companies create the dynamics of innovation* a model whereby the creation of knowledge is originated by the individual and, again through social processes, finalized by the collective.

These social interactions are coming to be evident in the academia and the research community, where interdisciplinary collaboration is becoming increasingly important to face big challenges. But besides understanding the social nature of knowledge itself, it is also essential to do it in a context in which technology and new forms of media, new distribution channels and new delivery systems are affecting traditional means of creating and disseminating knowledge. In particular, social networks and knowledge communities are specially reshaping the way knowledge is constructed by taking advantage of crowdsourcing. Hence, the importance of understanding how networks work and how knowledge emerges from their members' behaviour.

As information contained on social networks and online communities are generated by human beings, this is especially valuable for discovering knowledge about people's opinions and preferences, in addition to many other kinds of knowledge that we encode in text. Therefore, a strong background in text processing is necessary.

Social networks involves disciplines like sociology, mathematics, and computer science, whereas that text processing involves linguistics, statistics, and artificial intelligence. They are two good examples of interdisciplinary fields. For this reason, students from all backgrounds, but especially from the humanities, social science, computer science and engineering, are strongly encouraged to take this course.

###Methodology

This course is formed by 26 lectures over 13 weeks (two lectures per week). As it is a mostly practical, students will work on their computers.

The course aims to discover objective knowledge from online resources that supports future decisions by means of programming exercises focused on social media mining and natural language processing. To achieve these goals, the contents are divided into 4 modules:
1. Knowledge. This is an introduction to the course. It introduces some concepts on knowledge, emphasizing on its social nature.
2. Python. This module is a programming guide with Python that develops all the programming skills needed to face the rest of the course.
3. Texts. This module is the heart of the course. It covers subjects related to text manipulation, resource accessing and applied natural language processing.
4. Projects. Time for the students to work on their assignments, ask questions, and present their final projects.

All lectures are delivered in form of IPython Notebook and available through IPython Notebook viewer. They combine brief description of the theory with practical exercise to apply the concepts explained. The contents are taught in a broad and general way, but the examples and the assignments have been carefully designed to work with texts and contents on social media, as these are the main subjects of study in this course.

###Teaching philosophy

I have been a Computer Science lecturer and a Spanish instructor. From these very different experiences, I have drawn two conclusions:
1. Students are more productive when they enjoy their lectures.
2. Each student has his/her own process to acquire knowledge and develop his/her own inner skills.

To make lectures enjoyable, in-class learning must be a gratifying experience. I try to achieve this by building a good environment in the classroom and an approachable relationship with my students. I believe that my role in class must be to help them rather than to evaluate them. My formal evaluations as an instructor suggest that they perceive me as a friendly teacher, who creates a fun, comfortable, and stimulating learning atmosphere, and who portrays a committed attitude.

Based on the fact that each student has their own necessities and personal background and skills, my role as their instructor involves listening to them, understanding their needs, and working to meet these needs for each individual.

I do not think that classic memorization is as an effective way to learn. Instead, I prefer that my students learn by doing. For this reason, my lectures are focused, as much as possible, on the practical aspects of the topics taught. Inevitably, there always are lectures with a stronger theoretical viewpoint. But in general, I briefly describe the theory and then we see real applications of the concepts learned, giving a sense of utility to my lectures.

For the same reason, I do not evaluate my courses via exams, but rather by working on practical assignments. Rather, I evaluate their aptitude to solve assignments by allowing them to utilize all resources at hand. These resources vary from their own intellectual capabilities to resources in libraries or online instruments and their grade of involvement with the activities of course.
